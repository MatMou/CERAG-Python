# Anagrams 

```python
def char_counts(s):
    counts = {}
    
    for ch in s: 
        if ch in counts:
            counts[ch] = counts[ch]  + 1
        else: 
            counts[ch] = 1
    return counts 

def anagram(s1, s2):
    
    ct1 = char_counts(s1)
    ct2 = char_counts(s2)
    
    if ct1 == ct2:
        print("Yep. Anagram it is !  \n")
    else: 
        print("Nope ! \n")
```

```python
anagram('evil','live')
```

# Secrets messages and codes

```python
alphabet = [chr(i) for i in range(9,123)] #chr() of integer gives letter, ord() of character gives integer

def create_encoding(nb, alphabet):
    rd.seed(nb)
    encode = {}
    for i in alphabet: 
        encode[i] = str(rd.randrange(1,1000000))
    return encode

def encode_that(code, text):
    encoded_text = []
    for i in text:
        if code.get(i) != None:
            encoded_text.append(code.get(i))
        else : 
            encoded_text.append(str(i))

    coded_text = '-'.join(encoded_text)
    return coded_text

def decode_that(code, coded_text):
    decode  = {v: k for k, v in code.items()}
    split_text = coded_text.split('-')
    pre_final_text = []

    for i in split_text:
        if decode.get(i) != None:
            pre_final_text.append(decode.get(i))
        else: 
            pre_final_text.append(str(i))

    final_text = ''.join(pre_final_text)
    return final_text
```

```python
encode = create_encoding(18, alphabet)
text = "Hey there! Fancy seeing you there. Let's go to the party together. My number is: 2387192. Don't forget that my credit card number is 0987362573, and that the code is 8765. Also, my password for my bank account is 1267162. You know, just in case you need some money sometime"
text
```

```python
coded_text = encode_that(encode, text)
coded_text
```

```python
decoded_text = decode_that(encode, coded_text)
decoded_text 
```

```python
decoded_text == text
```
