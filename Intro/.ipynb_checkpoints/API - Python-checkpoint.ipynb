```python
import numpy as np 
import pandas as pd 
#import fix_yahoo_finance as fyf # only for python 2
import yfinance as fyf # only for python 3
from myFunctions import get_SP500tickers # only for python 3
import matplotlib.pyplot as plt 
from scipy.stats import norm 
from math import exp
```

We want to download 200 stocks from the SP500 index. There is two ways to do that. First, you go on a website that records all of the tickers of all the stocks and copy-paste them into a variable called `tickers`. Second option, you use fonctionalities of python in order to go on the wikipedia page of the SP500 and transform their table into a dataframe from which you just have to call the number of stocks you want. It is faster. 

In python 2, the function from `read_html()`, from the pandas package, is of great help and allow you to do this the fastest way possible. 

## Only for python 2

```python
# Python 2 
sp500_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data_table = pd.read_html(sp500_url)
tickers = data_table[0][1:][0].tolist() 
```

In python 3, the function `read_html()` might not work. You have a way around this using Web Scraping (which is outside the scope of this introduction). So, we have defined a user-defined function that is included in the module **myFunctions**. The function `get_SP500tickers()` is specifically designed to go and get all the tickers from the wikipedia page of the SP500. 

## Only for python 3

```python
tickers = get_SP500tickers()
```

Once we have the tickers we can specify that we only want 200 of them, and use the yahoo finance API to get the data. We use the function `download()` from the package **fix_yahoo_finance** (python 2) or **yfinance** (python 3), and use the nickname defined above **fyf**. Since, we are only interested by the value of the stock at the close of the day (adjusted) and the volume traded per day, we specify them at the end of the function. 

<br>

As a rule of thumbs, it is recommended to always take a look at how your data is. So we're using the function `head()` to print the first few rows. 

```python
stocks = tickers[0:200]
start = '2017-01-01'
end = '2018-01-01'
df = fyf.download(stocks,start, end)[['Adj Close', 'Volume']]
df.head()
```

Performing logarithmic return is then quite straightfroward. They are defined as: 

<br>
<center>$ R_t = ln(P_t/P_{t-1}) = ln(P_t) - ln(P_{t-1})  $</center> 
<br>

We create a dataframe in which we can store the result of the computation, using the **pandas** function `DataFrame()`. Here, the resulting dataframe is empty and will be filled directly with the ticker as column's names as we assign it the values of all log returns

<br>

We use the function `log()` from the **numpy** package (nicknamed **np**), and apply the differencing operator `.diff()`. The resulting `NA` needs to be moved out of the dataframe. We use the function `.dropna()`, and specify `how=all` to remove any columns having only `NA` values. Finally, we check if the data is as we wished. 

```python
Rt = pd.DataFrame()
Rt = np.log(df['Adj Close']).diff().dropna(how='all')
Rt.head()
```

A better way to check if your data makes sense is to plot it. In python, there are two main libraries for plotting: **plotly** and **matplotlib**. The first is great when it comes to 3D representation and interactive graphs, when the second is better for simple, yet elegant visualisations.  

<br>

We use here the **matplotlib** packages, with the extension **.pyplot** (nicknamed: **plt**). The first step in plotting is to create a blank canvas onto which you will plot your data. So, we initiate a blank object with `figure()` and specify the size of the graph (first number for length, second for height). We can then use `plot()` to plot the raw data onto the blank canvas. 

<br>

We could just stop there and specify `show()` in order to display the graph. But, we can easily make the plot more readable by specifying some optional functions. `legend()` gives out the legend and *loc* gives you the location of the legend (note that if the legend is too big, it will not be displayed). `grid(True)` makes the grid in the background appear. You specify the title and axis with `title()`, `xlabel()` and `ylabel()`. 

```python
#plot cumulative returns 
plt.figure(figsize=(15,7))
plt.plot(Rt.cumsum())
plt.legend(loc=2)
plt.grid(True)
plt.title('All stocks cumulative returns', fontdict={'fontsize':16})
plt.ylabel('Stocks Returns')
plt.xlabel('time')
plt.show()
```

Now that the data looks as expected, we can move on to the computation of the risk and liquidity indicators. Note that, since we used a dataframe (with **pandas** package), we can forget about any loop and just compute in one line the Value-at-Risk. Python will understand by itself that it is vector calculus and will return one value for each column. 

<br>

Reminder on VaR: 

<br>

<center> $VaR(q) \equiv Pr(R_t \leq VaR(q)) = q$ </center>

<br>

Hence, assuming gaussian return's distribution: 

<br>

<center> $VaR(q) = \mu*n + \sigma*\sqrt{n}*\Phi^{-1}(q)$ </center>



```python
#Value-at-Risk
Invest =1.
c = 0.025
n = 1
VaR = Invest*(np.mean(Rt)*n + np.std(Rt)*np.sqrt(n)*norm.ppf(c))
VaR
```

Same for the liquidity ratio, we take the illiquidity proxy of Amihud (2012): 

<br>

<center> $ Liquidity = \frac{1}{N} \sum \frac{|R_{i,t}|}{\$ Vol_{i,t}} $ </center> 


With i going from 1 to N. 

```python
#Liquidity Ratio 
USD_volume = df['Volume']*df['Adj Close']
Liquidity = np.mean(abs(Rt)/USD_volume)
Liquidity
```

We create a new dataframe to store the value of both of our measurement. 

```python
indics = pd.DataFrame()
indics['Liquidity'] = Liquidity
indics['VaR'] = VaR
indics = indics.fillna(0)
indics.head()
```

We come back to plotting our results. Same as before, we used the `figure()` function of the **matplotlib.pyplot** in order to create a blank canvas. Since, an appropriate visualisation for 1 value per firm would be a barplot, we use the function `bar()` specifying first the xaxis and then the actual data. Since, there would be an overflow of data in the graph, and we want to keep it human-readable, we only plot 50 firms and we rotate the name in the x axis using the `xticks()` function and specifying the rotation angle. 

```python
plt.figure(figsize=(12,7))
plt.bar(indics.index[0:50], indics['Liquidity'][0:50])
plt.grid(True)
plt.xticks(rotation=45)
plt.ylabel('ILLIQ')
plt.xlabel('Firm')
plt.title('Liquidity index', fontdict={'fontsize':22})
plt.show()
```

The color by default is blue, but we can also change the color by specifying it in the `color` parameter. 

```python
plt.figure(figsize=(12,7))
plt.bar(indics.index[60:80], indics['VaR'][60:80], color='darkred')
plt.grid(True)
plt.xticks(rotation=45)
plt.ylabel('VaR')
plt.xlabel('Firm')
plt.title('Value at Risk, daily', fontdict={'fontsize':22})
plt.show()
```

The use of **colormaps** can also be interesting in some cases. A colormap is a list of color, which your then call inside the plotting specification. 

```python
#adding some colors to the plot 
cmap = plt.get_cmap('gnuplot')
colors = [cmap(i) for i in np.linspace(0, 1, 20)]

plt.figure(figsize=(12,7))
plt.bar(indics.index[60:80], indics['VaR'][60:80], color=colors)
plt.grid(True)
plt.xticks(rotation=45)
plt.ylabel('VaR')
plt.xlabel('Firm')
plt.title('Value at Risk, daily', fontdict={'fontsize':22})
plt.show()
```

Sometimes, presenting results may need multiple graphs on  single page. It allows for an easier comparability of the data and summarize results, often better than a long paragraph. After creating your blank canvas, you can use the fonction `subplot()` to separate the canvas in few parts. The first two numbers represents how you divide the canvas (here 2 columns and 2 rows). The third give the place of the plot on the canvas. 

```python
plt.figure(figsize=(17,10))
#barplot 
plt.subplot(2,2,1)
plt.bar(indics.index[20:50], indics['VaR'][20:50])
plt.xticks(rotation=45)
plt.title('Value at risk', fontdict={'fontsize':16})

#lines
plt.subplot(2,2,2)
plt.bar(indics.index[20:50], indics['Liquidity'][20:50])
plt.xticks(rotation=45)
plt.title('Liquidity Index', fontdict={'fontsize':16})

#histogram
plt.subplot(2,2,3)
plt.hist(indics['VaR'])
plt.title('Empirical Dist of VaR', fontdict={'fontsize':16})

#scatter 
plt.subplot(2,2,4)
plt.scatter(indics['VaR'],indics['Liquidity'])
plt.title('Scatter VaR-Liquidity', fontdict={'fontsize':16})
plt.show()
```
